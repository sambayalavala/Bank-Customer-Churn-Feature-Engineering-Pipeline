```python
from pyspark.sql.functions import col

#Row counts for each layer
bronze = spark.table("churn_catalog.raw.customer_data").count()
silver = spark.table("churn_catalog.processed.customer_profiles").count()
gold   = spark.table("churn_catalog.analytics.churn_features").count()
ml     = spark.table("churn_catalog.analytics.churn_ml_ready").count()

counts_df = spark.createDataFrame(
    [(bronze, silver, gold, ml)],
    ["bronze_rows", "silver_rows", "gold_rows", "ml_rows"]
)

print("Row counts by layer:")
display(counts_df)

# Last 20 data-quality issues from logs table
logs_df = (
    spark.table("churn_catalog.logs.data_quality")
         .orderBy(col("timestamp").desc())
         .limit(20)
)

print("Recent data-quality issues:")
display(logs_df)
```
