```python
from pyspark.sql.functions import col

# Load GOLD table
df = spark.table("churn_catalog.analytics.churn_features")

#TEST 1: churn must not be null 
null_churn = df.filter(col("churn").isNull()).count()
assert null_churn == 0, f" TEST FAILED: churn has {null_churn} null rows"

# TEST 2: age must be between 1 and 80
bad_age = df.filter((col("age") <= 0) | (col("age") >= 80)).count()
assert bad_age == 0, f"TEST FAILED: {bad_age} rows with invalid age"

#  TEST 3: balance cannot be negative 
neg_balance = df.filter(col("balance") < 0).count()
assert neg_balance == 0, f" TEST FAILED: {neg_balance} rows with negative balance"

print(" All KPI tests passed successfully!")
```
